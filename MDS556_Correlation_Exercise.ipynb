# Correlation Exercise

```python
# importing libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# importing data and storing as a dataframe

path = r"https://raw.githubusercontent.com/crosstrack10/data-science-projects/main/datasets/Random%20data%20v2.csv"
randomdata = pd.read_csv(path)
```

```python
# returning 10 random records

randomdata.sample(10)
```

```python
# printing correlation

print (randomdata.corr())
```

```python
# plotting correlation heatmap

plt.figure(figsize=(7,7))
sns.heatmap(randomdata.corr(),
            vmin=-1,
            cmap=sns.diverging_palette(20, 220, n=200));
```

The heatmap shows most variables have weak relationships with other variables. There are a few exceptions, but at best, they seem to have moderate relationships.

```python
# importing data and storing as a dataframe

path = r"https://raw.githubusercontent.com/crosstrack10/data-science-projects/main/datasets/Random%20data%20v2tr.csv"
randomdatatr = pd.read_csv(path)
```

```python
# returning 10 random records

randomdatatr.sample(10)
```

```python
# printing correlation

print(randomdatatr.corr())
```

```python
# plotting correlation heatmap

plt.figure(figsize=(7,7))
sns.heatmap(randomdatatr.corr(),
            vmin=-1,
            cmap=sns.diverging_palette(20, 220, n=200));
```

There are so many variables that it makes it difficult to see relationships between them. The correlation histogram should provide better insights.

```python
# plotting histogram for better insights

plt.style.use('ggplot')
plt.figure(figsize=(12,6))
ax = plt.hist(randomdata.corr())
plt.xlabel('correlation')
plt.ylabel('count');
```

The view does not change my interpretation of the previous heatmap.

```python
# plotting histogram for better insights

plt.figure(figsize=(12,6))
ax = plt.hist(randomdatatr.corr())
plt.xlabel('correlation')
plt.ylabel('count');
```

Although there are still many variables with weak correlation values, many have moderate to moderately strong values (0.3 - 0.6). In addition, the correlation frequency distribution is normally distributed. In this case, having fewer features with more data points improves the correlation values.
